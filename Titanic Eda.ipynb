The Titanic dataset consists of **891 passenger records** and **12 features**, providing detailed information about passenger demographics, ticket details, travel class, fare, and survival status. It contains both **numerical** and **categorical** variables, making it an excellent dataset for Exploratory Data Analysis (EDA) and machine learning.

### Key Findings
- The dataset contains **891 rows** and **12 columns**.
- It includes a combination of numerical and categorical features.
- Missing values are present in:
  - **Age:** 177 missing values (~19.9%)
  - **Cabin:** 687 missing values (~77.1%)
  - **Embarked:** 2 missing values (~0.2%)
- The **Cabin** column has a significant amount of missing data and may require feature engineering or removal.
- The **Age** column requires imputation before analysis or modeling.
- The **Embarked** column has only two missing values, which can be filled using the most frequent category.

### Overall Assessment
The dataset is well-structured and suitable for data analysis. It provides enough information to explore how passenger characteristics such as **gender, age, passenger class, fare, family size, and embarkation port** influenced survival during the Titanic disaster.

Before performing detailed EDA or building predictive models, the dataset should undergo:
1. Missing value treatment.
2. Duplicate value check.
3. Outlier detection.
4. Feature engineering (e.g., Family Size, Title, Deck).
5. Data encoding and preprocessing where necessary.

Overall, the Titanic dataset is a clean and beginner-friendly dataset that serves as an excellent example for practicing data cleaning, visualization, feature engineering, and predictive modeling.

# Feature Engineering 

```python
df["Title"] = df["Name"].str.extract(r',\s*([^\.]+)\.')
df['Title'].value_counts()
```

```python
df["Title"] = df["Title"].replace({
    "Mlle": "Miss",
    "Ms": "Miss",
    "Mme": "Mrs"
})
df['Title'].value_counts()
```

```python
def modify_title(title):
    common_titles = ["Mr", "Miss", "Mrs", "Master"]

    if title in common_titles:
        return title
    else:
        return "Rare"
```

```python
df["Title"] = df["Title"].apply(modify_title)
df['Title'].value_counts()
```

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1
df['FamilySize']
```

```python
def family_type(size):
    if size == 1:
        return "Alone"
    elif 2 <= size <= 4:
        return "Small Family"
    else:
        return "Large Family"
```
